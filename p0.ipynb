```python
# WAP to exatract all the upper case alphabate from given string.

a='PyTHON'
i = 0
out = ''
while i < len(a):
    if 'A' <= a[i] <= 'Z':
        out += a[i]
    i += 1
print(out)
```

```python
# WAP to extract all the vowels from the string.

a = 'Elephant'
count = 0
i = 0
out = ''
while i < len(a):
    if a[i] in 'AEIOUaeiou':
        out += a[i]
        count += 1
    i += 1
print("count :", count)
print("output :",out)
```
